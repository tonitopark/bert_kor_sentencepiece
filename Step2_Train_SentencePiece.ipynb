```python
import sentencepiece as sp
import glob
```

```python
prefix = "model/sentencepiece/wiki-ko"
vocab_size = 32000
ctl_symbols = '[PAD],[CLS],[SEP],[MASK]'

file_list = glob.glob(f'data/wiki-kor/**/*.txt')
files = ','.join(file_list)
print(files)
```

```python
command = f"""--input={files} --model_prefix={prefix}
             --vocab_size={vocab_size} --control_symbols={ctl_symbols}"""
print(command)
sp.SentencePieceTrainer.Train(command)
```

### How to load vocabulary

```python
import collections

# Load  vocabulary file into a dictionary.
vocab = collections.OrderedDict()
index = 0
with open("model/sentencepiece/wiki-ko.vocab", "r") as reader:
    while True:
        token = reader.readline()
        if not token:
            break
        token, _ = token.split("\t")
        token = token.strip()
        vocab[token] = index
        index += 1
```
